```python
import pandas as pd #importing pandas 
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe() #gives summary stastistics of only numerical columns
```

```python
df.describe(include='all')#gives summary statistics of all columns
```

```python
df.isnull().sum()#finding missing values
```

```python
df['Review Rating']= df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()#converting all the col to lower case
```

```python
df.columns
```

```python
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a new column age_group
labels=['Young Adult','Adult','Middle Aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency_days
frequency_mapping = {
       'Fortnightly': 14,
       'Weekly': 7,
        'Monthly': 30,
        'Quarterly': 90,
        'Bi-weekly': 14,
        'Anually': 365,
        'Every 3 Months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df['purchase_frequency_days'].dtype
```

```python
df['purchase_frequency_days'] = pd.to_numeric(
    df['purchase_frequency_days'],
    errors='coerce'
).astype('Int64')
```

```python
df['purchase_frequency_days'].dtype
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied']==df['promo_code_used']).all()#checking whether both the columns are same
```

```python
df=df.drop('promo_code_used',axis=1)
```

```python
from sqlalchemy import create_engine

#step 1: Connect to postregesql
#replace placeholders with your actual details
username = "postgres"  #default user
password = "root" #the password you set during installation
host = "localhost"   #if running locally
port = "5432"  #default postgresql port
database = "customer_behaviour"   #the database we created in the pgadmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

#step 2:load dataframe into postgresql
table_name = "customer"   #choose any table name
df.to_sql(table_name,engine, if_exists="replace", index=False)

print(f"Data sucessfully loaded into table '{table_name}' in database '{database}'.")
```
